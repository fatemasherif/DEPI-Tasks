```python
import pandas as pd
import numpy as np
import sqlite3
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\fatem\Downloads\Titanic.csv")
df.head(10)
```

#Bar Graph

```python
survivalcounts=df['Survived'].value_counts()
survivalcounts.plot(kind='bar')
plt.title('Survival passengers')
plt.xlabel('Survived')
plt.ylabel('No of passengers')
plt.show()
```

```python
survbyclass=df.groupby('Pclass')['Survived'].mean()
survbyclass.plot(kind='bar')
plt.title('Survival by each class')
plt.xticks([0, 1, 2], labels=["1st Class", "2nd Class", "3rd Class"], rotation=0)
plt.show()
```

#Histograms

```python
plt.hist(df[df['Survived']==1]['Age'], bins=20,alpha=0.5, label='Survived', color='green')
plt.hist(df[df['Survived']==0]['Age'], bins=20, alpha=0.5, label='Didnt survived', color='red')
plt.title('Age distribution of passengers')
plt.xlabel('Age')
plt.ylabel('Number of passengers')
plt.show()
```

#Scatter Plot

```python
survbysex=df.groupby('Sex')['Survived'].sum()
plt.scatter(survbysex.index,survbysex.values, color=['pink', 'blue'], s=300)
plt.title('Number of survivors by gender')
plt.show()
```

#Line

```python
survbyclass=df.groupby('Pclass')['Survived'].mean()
plt.plot(survbyclass.index, survbyclass.values)
plt.title("Survival Rate by Passenger Class")
plt.xlabel('Pasenger class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
survbyclass=df.groupby('Pclass')['Survived'].mean()
survbyclass.plot(kind='line')
plt.title("Survival Rate by Passenger Class")
plt.xlabel('Pasenger class')
plt.ylabel('Survival Rate')
plt.show()
```

#Boxplot

```python
sns.boxplot(x='Pclass', y='Fare', data=df, palette='viridis')
plt.title("Passengers distribution by passenger class")
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```
